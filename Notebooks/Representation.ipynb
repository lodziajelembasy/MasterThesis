## Wektoryzacja z użyciem metody embedding
### Dokument jako wektor
### Doc2vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tagged_data = [TaggedDocument(words=d, tags=[i]) for i, d in enumerate(df.iloc[:,-1])]
print(tagged_data[0])
```

#### Uczenie na piosenkach

```python
model = Doc2Vec(dm=1)
model.build_vocab(tagged_data)
model.train(tagged_data,
           total_examples=model.corpus_count,
           epochs=model.epochs)
model.delete_temporary_training_data(keep_doctags_vectors=True,
                                     keep_inference=True)
```

```python
df_doc2vec = pd.DataFrame(model.docvecs.vectors_docs, index=df.index)
df_doc2vec.head()
```

```python
df_doc2vec.to_csv('../CSV/piosenki_doc2vec_default_dm1_lem_pl', index=True)
```

```python
readings_vectors = pd.DataFrame()
for i in range(len(df2)):
    y = model.infer_vector(df2.iloc[i,-1])
    readings_vectors[i] = y
print(readings_vectors.shape)
#readings_vectors.head()
```

```python
df2_doc2vec = pd.DataFrame(np.transpose(readings_vectors.values),
                           index=df2.index)
df2_doc2vec.head()
```

```python
df2_doc2vec.to_csv('../CSV/czytania_doc2vec_default_dm1_lem_pl', index=True)
```
