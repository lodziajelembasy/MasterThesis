# Word2vec for cosine similarity

```python
import pandas as pd 
from gensim.utils import tokenize
import numpy as np
```

```python
df = pd.read_csv('../CSV/prep_songs_pl_lem_stop_wcrft1.csv', header=0, index_col=None)
df = df.dropna()
df2 = pd.read_csv('../CSV/prep_readings_pl_lem_stop_wcrft1.csv', header=0, index_col=None)
print(df.shape)
print(df2.shape)
df.head(2)
```

```python
df['title_tokenized'] = df['title'].apply(lambda x: list(tokenize(x, to_lower=True)))
df['title_cleaned'] = df['title_tokenized'].apply(lambda x: " ".join([z.lower() for z in x if z.isalpha()]))

print(df.shape)
df.head()
```

```python
df.index = df.title_cleaned.values
df = df.drop(['title_cleaned', 'title_tokenized'], axis=1)
df.head()
```

```python
df2.index = df2.date.values
df2 = df2.drop(['date'], axis=1)
df2.head()
```

```python
from gensim.models.keyedvectors import KeyedVectors
model_path = 'PATH\\nkjp-lemmas-all-300-skipg-ns.txt'
w2v_model = KeyedVectors.load_word2vec_format(model_path, binary=False)
```

```python
w2v_model.get_vector('słoń')
```

```python
def document_vector(word2vec_model, doc):
    doc = [word for word in doc if word in word2vec_model.vocab]
    return np.mean(word2vec_model[doc], axis=0)

from ast import literal_eval 
  
def convert(lst): 
    return literal_eval(lst) 
```

```python
#document_vector(w2v_model, convert(df['txt_pl_lem_tokenized'][0]))
```

```python
avg_w2v_songs = df['txt_pl_lem_tokenized'].apply(lambda x: pd.Series(document_vector(w2v_model, convert(x))))
avg_w2v_songs.head()
```

```python
#avg_w2v_songs.to_csv('../CSV/piosenki_avg_w2v_nkjp-lemmas-all-300-skipg-ns_pl_lem.csv',header=True, index=True)
```

```python
avg_w2v_readings = df2['txt_pl_lem_tokenized'].apply(lambda x: pd.Series(document_vector(w2v_model, x)))
avg_w2v_readings.head()
```
